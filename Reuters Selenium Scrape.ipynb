```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import os
import time
```

```python
chromedriver = "/Applications/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
URL = "https://www.reuters.com/news/technology"

driver = webdriver.Chrome(chromedriver)
driver.get(URL)
```

```python
documents = {}
num_elements = 1
time.sleep(10)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
while True:
    time.sleep(10)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    items = driver.find_elements_by_class_name('FeedItem_item_AanJv')
    if len(items) == num_elements:
        print('Break')
        break

    prev_num_elements = num_elements
    num_elements = len(items)
    print("Number of Elements: %d" % num_elements)
    for item in items[prev_num_elements-1:]:
        title = item.find_element_by_tag_name('h2').text
        if title not in documents.keys():
            documents[title] = {}
            documents[title]['link'] = item.find_element_by_tag_name('h2').find_element_by_tag_name('a').get_attribute('href')
            documents[title]['date_updated'] = item.find_element_by_class_name('FeedItemMeta_date-updated_1EZPz').text
            documents[title]['blurb'] = item.find_element_by_class_name('FeedItemLede_lede_Wa-ek').text
```

```python
new_docs = []
for key in documents.keys():
    new_docs.append({
        'title': key,
        'date_updated': documents[key]['date_updated'],
        'link': documents[key]['link'],
        'blurb': documents[key]['blurb']
    })
```

```python
import json
with open('reuters_tech.json', 'w') as fout:
    json.dump(new_docs, fout)
```
